```python
import numpy as np
from sklearn.linear_model import LinearRegression
%matplotlib inline
X = np.arange(3,70)
X = X.reshape(-1,1)
y = np.dot(X, np.array([1.5]))**1 + np.random.normal(0,10,len(X))
reg = LinearRegression().fit(X, y)
import matplotlib.pyplot as plt
```

Plotting the scatter and finding regression score

```python
plt.plot(X,y,'o')
plt.plot(X,np.dot(X,reg.coef_)+reg.intercept_)
reg.score(X, y)
```

Importing real Climate Change Data

```python
import pandas as pd
df = pd.read_csv("climate_change.csv")
```

Split data into training (70%) and testing

```python
X = pd.DataFrame(df['CO2'])
y = pd.DataFrame(df['Temp'])
X_train = X[0:214]
y_train = y[0:214]

X_test = X[214:]
y_test = y[214:]
```

Plotting data and fitting Linear Regression using scikit learn. This is entirely for training period

```python

fig, ax = plt.subplots(figsize=(20,10))

# We change the fontsize of minor ticks label 
ax.tick_params(axis='both', which='major', labelsize=18)
ax.tick_params(axis='both', which='minor', labelsize=16)
ax.plot(X_train, y_train,'o')
plt.xlabel("CO2", fontsize = 18)
plt.ylabel("Global Temperature", fontsize = 18)
plt.title("Relationship between Global temperature and CO2", fontsize = 30)


reg = LinearRegression().fit(X_train, y_train)

ax.plot(X_train,np.dot(X_train,reg.coef_)+reg.intercept_, linewidth = 8)
```

Testing model performance on test data

```python
fig, ax = plt.subplots(figsize=(20,10))

# We change the fontsize of minor ticks label 
ax.tick_params(axis='both', which='major', labelsize=18)
ax.tick_params(axis='both', which='minor', labelsize=16)
ax.plot(X_test, y_test,'o',color ="red")
ax.plot(X_test,np.dot(X_test,reg.coef_)+reg.intercept_,'o')


plt.xlabel("CO2", fontsize = 18)
plt.ylabel("Global Temperature", fontsize = 18)
plt.title("Relationship between Global temperature and CO2", fontsize = 30)

plt.legend(["Ground Truth","Predictions"], loc = 8, fontsize = 30)

reg = LinearRegression().fit(X_train, y_train)

ax.plot(X_train,np.dot(X_train,reg.coef_)+reg.intercept_, linewidth = 8)


plt.show()
```

Direct command to predict data for input

```python
reg.predict(X_test)
```

Setting up Multiple Regression Problem

```python
X = pd.DataFrame(df)
X = X.drop(columns = ['Temp'])
y = pd.DataFrame(df['Temp'])
X_train = X[0:214]
y_train = y[0:214]

X_test = X[214:]
y_test = y[214:]
```

Fitting multivariate linear regression relationship

```python
reg = LinearRegression().fit(X_train, y_train)
```

Performance score for Training

```python
reg.score(X_train, y_train)
```

Performance score for Testing. This is R^2 score.

```python
reg.score(X_test, y_test)
```
